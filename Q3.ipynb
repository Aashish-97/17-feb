```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (Ensure you replace this with your actual dataset)
# data = pd.read_csv('your_data.csv')

# Analyze host listings distribution
data = pd.read_csv('/content/listings_detailed.csv')
host_listings = data.groupby('host_id')['id'].count().sort_values(ascending=False)

# Calculate statistics
total_hosts = len(host_listings)
multiple_property_hosts = len(host_listings[host_listings > 1])
percentage_multiple = round((multiple_property_hosts / total_hosts * 100), 2)

# Print statistics
print(f"\nHost Property Analysis:")
print(f"Total number of hosts: {total_hosts}")
print(f"Hosts with multiple properties: {multiple_property_hosts} ({percentage_multiple}%)")

# Extract top 5 hosts
top_hosts = host_listings.head(5)

# Get host names for visualization
top_host_names = [data[data['host_id'] == host_id]['host_name'].iloc[0] for host_id in top_hosts.index]

# Bar chart: Top 5 hosts by number of properties
plt.figure(figsize=(10, 5))
plt.barh(top_host_names[::-1], top_hosts[::-1], color='skyblue')
plt.xlabel("Number of Properties")
plt.ylabel("Host Names")
plt.title("Top 5 Hosts by Number of Properties")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Additional analysis of property distribution
property_counts = host_listings.value_counts().sort_index()

# Histogram: Distribution of properties per host
plt.figure(figsize=(10, 5))
plt.bar(property_counts.index, property_counts.values, color='coral')
plt.xlabel("Number of Properties Owned")
plt.ylabel("Number of Hosts")
plt.title("Distribution of Properties per Host")
plt.yscale("log")  # Log scale for better visualization if counts vary widely
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
